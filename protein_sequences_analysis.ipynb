```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import pandas as pd
from Bio import SeqIO
from Bio.SeqUtils import GC
```

```python
# Build Protein Distrution
def get_protein_distribution(fibrinogen_protein_types):
    # Sequence Analysis of Proteins
    # Protein sequencing is the method of determining the amino acid sequence of all or part of a protein or peptide.
    # Files are assumed to be in .fasta format 
    my_dictionary  = {}
    amino_acids = {}
    for fibrinogen_protein_type in fibrinogen_protein_types:
        print(f"----------------------{fibrinogen_protein_type}--------------------------------")
        for seq_record in SeqIO.parse("DataSources/" + fibrinogen_protein_type, "fasta"):
            #Amino Acids
            amino_acids =  dict({
                "A" : seq_record.seq.count('A'),
                "B" : seq_record.seq.count('B'),
                "C" : seq_record.seq.count('C'),
                "D" : seq_record.seq.count('D'),
                "E" : seq_record.seq.count('E'),
                "F" : seq_record.seq.count('F'),
                "G" : seq_record.seq.count('G'),
                "H" : seq_record.seq.count('H'),
                "I" : seq_record.seq.count('I'),
                "K" : seq_record.seq.count('K'),
                "L" : seq_record.seq.count('L'),
                "M" : seq_record.seq.count('M'),
                "N" : seq_record.seq.count('N'),
                "P" : seq_record.seq.count('P'), 
                "Q" : seq_record.seq.count('Q'), 
                "R" : seq_record.seq.count('R'),
                "S" : seq_record.seq.count('S'),
                "T" : seq_record.seq.count('T'),
                "U" : seq_record.seq.count('U'),
                "V" : seq_record.seq.count('V'),
                "W" : seq_record.seq.count('W'),
                "Y" : seq_record.seq.count('Y'),
                "Z" : seq_record.seq.count('Z'),
                "X" : seq_record.seq.count('X')
            })

            if 'phenylalanine' in my_dictionary:
                pitems = my_dictionary['phenylalanine']
                pitems.append(amino_acids['F'])
                my_dictionary['phenylalanine'] = pitems 
                mitems = my_dictionary['methionine']
                mitems.append(amino_acids['M'])
                my_dictionary['methionine'] = mitems
                sitems = my_dictionary['serine']
                sitems.append(amino_acids['S'])
                my_dictionary['serine'] = sitems
            else:
                my_dictionary = {'phenylalanine' : [amino_acids['F']], 'methionine': [amino_acids['M']],  'serine': [amino_acids['S']]}
    return my_dictionary
```

```python
# Plot FGA Distribution
fibrinogen_protein_types = ["FGA_proteins.fasta"]
fga_distributions = get_protein_distribution(fibrinogen_protein_types)
fga = pd.DataFrame({'phenylalanine': fga_distributions['phenylalanine'],
                         'methionine': fga_distributions['methionine'],
                         'serine': fga_distributions['serine'],
                        },
                        index=['NP_068657.1', 'NP_000499.1', 'XP_054205235.1'])

ax = fga.plot(kind='bar', stacked=True, color=['green', 'blue', 'red'])

# Set Data Labels
ax.bar_label(ax.containers[0], label_type='center', color= 'white')
ax.bar_label(ax.containers[1], label_type='center', color= 'white')
ax.bar_label(ax.containers[2], label_type='center', color= 'white')

# Add Title and Labels
plt.title('FGA - Amino Acids Distribution')
plt.xlabel('Isoforms')
plt.ylabel('amino acids levels')
```
